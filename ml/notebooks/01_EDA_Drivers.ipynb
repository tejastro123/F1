# Exploratory Data Analysis: F1 Driver Performance
Analyzing historical trends and feature correlations.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_parquet('../../data/processed/f1_features_full.parquet')
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Matrix Matrix')
plt.show()
```
