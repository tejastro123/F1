# Feature Engineering Pipeline
Transforming raw Ergast/OpenF1 data into ML-ready features.

```python
import pandas as pd
import numpy as np

def calculate_form_index(results_df):
    # Rolling average of last 5 races
    return results_df.groupby('driverId')['position'].transform(lambda x: x.rolling(5, min_periods=1).mean())

print("Feature engineering logic documented.")
```
