# Monte Carlo Simulation: 2026 Championship Forecast
Running 10,000 simulations to determine the most likely champion.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

drivers = ['VER', 'HAM', 'LEC', 'NOR']
probs = [0.4, 0.2, 0.15, 0.25]

sims = np.random.choice(drivers, size=10000, p=probs)
pd.Series(sims).value_counts().plot(kind='bar', color='red')
plt.title('Monte Carlo Championship Distribution')
plt.show()
```
